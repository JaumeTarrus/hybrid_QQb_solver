# Schrödinger eq. solvers for Hybrid quarkonium

In this nootebook the hybrid quarkonium spectrum associated to the $\Sigma_u^-$ and $\Pi_u$ static energies is obtained. Results based on a earlier version of this nootebook where presented in Phys.Rev.D 92 (2015) 11, 114019 (arXiv: 1510.04299). See this reference for further details. 

For each angular momentum, exept 0, value we have two possible solutions corresponding to the coupled and uncoupled equations in eq.51 and 52 arXiv:1510.04299. This corresponds to wave functions in C.21 and C.22 respectively. To distinguish the two type of solution, we label them with 'p' and 'm'. The l=0 case is a special case of the former and we label it also with a 'p'.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp, simpson  
from scipy.interpolate import interp1d
```

```python
sec = 'bb' # use cc or bb for charmonium and bottomonium respectively
hbc = 0.1973269788
mc = 1.477 # charm mass in RS' scheme at nu_f=1Gev
mb = 4.863 # bottom mass in RS' scheme at nu_f=1Gev

sp_tab = np.loadtxt('V_1pm_tab.txt')      # Load tabulated potential in GeV units
pot_Sigma = interp1d(sp_tab[:,0],sp_tab[:,1]) # potential function generated interpolating the data
pot_Pi = interp1d(sp_tab[:,0],sp_tab[:,2]) # potential function generated interpolating the data

# The potential is slightly different from the one in 1510.04299 since the fit used here is constrained to become r*(0.21Gev^2)
# in the large r limit

def sctsl(vcc,vbb):  # this function selects values that are different for the two heavy quark sectors
    if sec == 'cc': aux = vcc
    elif sec == 'bb': aux = vbb
    return aux

# Uncomment the lines below to plot the two potentials

#x = sp_tab[:,0]
#y1 = sp_tab[:,1]
#y2 = sp_tab[:,2]

#plt.ylim(0, 8)
#plt.plot(x,y1,x,y2)
#plt.show()
```

In the cell below the coupled $\Sigma-\Pi$ case is solved (eq.51 in arXiv: 1510.04299). The strategy is analogous to the one used in the signle channel case for the quarkonium spectrum. That is, we scan upwards the values of the energies until we cross an eigenvalue, then the go back one step  and continue the scan with steps one order of magnitude smaller. This loop is stoped when a the wave functions show sufficient convergence to zero in the large r or when a certain accuracy on the eigenenergy is achieved. In the coupled equation case instead of the single radial wave fucntion as in quarkinium we look at the determinant of the two pairs on independent solutions of the coupled equation system (see arXiv: 1510.04299 for the details).

Note that in this notebook we use the standard definition $\hat{\bf{r}}_+=-(\hat{\bf{\theta}}\pm\hat{\bf{\phi}})/\sqrt{2}$ which differs by a minus sign with the one in arXiv:1510.04299. This results in different sign in front of the off-diagonal terms in the coupled Schrödinger equations and a corresponding change in the initial conditions of the wave functions.

```python
mq = sctsl(mc, mb) # picks heavy quark mass 

# Each step in the loop increses the energy tested by stp*10**(-sf)

stp = 1 # order 1 factor multipling the energy step
sf = 1  # scale factor controling the order of magnitude of the initial steps (now set to 0.1)
sd = 10  # significant digits in the final results of the energies (more or less...). More is needed for larger maxr
spec = []

coupling_label = 'p' # This is labeling that this solutions correspond to C21 arXiv: 1510.04299. 
# p=plus which is the relative sign of the two \Pi terms with the r_+ definition we are using here

# Note that l>=1 for the coupled case
for l in range(1,3): # We loop over angular momentum 1,2; to be adjusted as needed

#   Starting energy for the scan. Since the potentials have a minumum we use that. Pi_u is always lower than \Sigma_u^- 

    en = np.min(sp_tab[:,2])
    
    for n in range(1,6):  # We loop over principal quarntum number n=1..8; to be adjusted as needed
        
        # Minimum value of the range of r for which we will solve the Schroedinger eq. Must be larger than r of the first data 
        # point of the potential.
            
        minr = 10**(-2)/hbc
        
        # Maximum value of the range of r. Ideally the larger the better, however larg values result in the need to find the
        # eigenenergies with unpractical accuracies. The most practial max r is just a few tens of a fm beyond where the wave 
        # function becomes almost zero, which I determine by trial an error and cheking the resulting wave functions. For 
        # increasing n and l we expect the extension of the wave function to increase. However, the scalings are also obtained  
        # by trial an error (maybe exact solutions of linear potentials could be used as a guide too?).
        
        maxr = sctsl(2.05*(1.10**(n-1))*(1.10**l)/hbc,1.5*(1.1**(n-1))*(1.08**l)/hbc)
        
        r_range =  [minr, maxr]
        
        # Boundary conditions of the r.w.f. and its derivative. These are obtained from the solution for the schroedinger
        # eq. in the short distance
        
        y_minr_p = [-np.sqrt(l)*minr**l,-(np.sqrt(l)**3)*minr**(l-1),np.sqrt(l+1)*minr**l,np.sqrt(l+1)*l*minr**(l-1)]
        y_minr_m = [np.sqrt(l+1)*minr**(l+2),np.sqrt(l+1)*(l+2)*minr**(l+1),np.sqrt(l)*minr**(l+2),np.sqrt(l)*(l+2)*minr**(l+1)]

        # We rewrite the schroedinger eq. into a linear system y'=func(r,y)
        # y[0]= psi_Sigma(r), y[2]= psi_Pi(r) (*reduced* w.f.) y[1]= psi_Sigma(r)' y[3]= psi_Pi(r)'
        
        def func(r,y):
            return [y[1],
                    mq*(pot_Sigma(r)-en)*y[0]+y[0]*(l*(l+1)+2)/r**2-(2*y[2]*(l*(l+1))**(1/2))/r**2,
                    y[3],
                    mq*(pot_Pi(r)-en)*y[2]+y[2]*l*(l+1)/r**2-(2*y[0]*(l*(l+1))**(1/2))/r**2]   
                 
        # we solve the coupled diff. eq. for the initial value and find the sign of the divergence
        
        in_psi_p = solve_ivp(func, r_range, y_minr_p, rtol=10**(-sd-1), atol=10**(-sd-1)) 
        in_psi_m = solve_ivp(func, r_range, y_minr_m, t_eval=in_psi_p.t, rtol=10**(-sd-1), atol=10**(-sd-1)) 
        
        ref_sgn = np.sign(in_psi_p.y[0][-1]*in_psi_m.y[2][-1]-in_psi_p.y[2][-1]*in_psi_m.y[0][-1])
#              
        # Now we start the loop that scans the energy (from the initial value and incresing) until it finds an eigenenergy up to
        # some accuracy
        
        while True:
            en = en + stp*10**(-sf) # Adding energy step

            psi_p = solve_ivp(func,r_range,y_minr_p, rtol=10**(-sd-1), atol=10**(-sd-1)) # Solve diff.eq. for the energy of the step
            psi_m = solve_ivp(func,r_range,y_minr_m, t_eval=psi_p.t, rtol=10**(-sd-1), atol=10**(-sd-1)) # Solve diff.eq. for the energy of the step
            
            det_psi = psi_p.y[0]*psi_m.y[2]-psi_p.y[2]*psi_m.y[0]
            
            # value of the deterinant of the solutions of the w.fs at the edge of the range over it max.value
            
            ref = np.absolute(det_psi[-1])/np.nanmax(np.absolute(det_psi)) 
            
            sgn = np.sign(det_psi[-2]) # sign of the determinant at the edge of the r range

            if (ref < 1e-3): # We approximate the eigenenergy by the value of en when ref is this small
                break      # We exit the loop
            if sf == sd:   # We exit the loop if the number of significat digits of the energy becomes too large
                print('Precision reached') # if this prints the convergences conditions are too harsh
                break
            if sgn != ref_sgn: # If the sign of the large r divergence changes means we have crossed an eigenvalue
                en = en - stp*10**(-sf) # Undo the last energy step
                sf=sf+1 # We increase the scale factor so the next energy step is one order of magnitude smaller.
        
        # We compute the number of nodes by counting adjacent points in the determinant with different sign
        
        nodes = np.sum(det_psi[0:-100]*det_psi[1:-99] < 0) 
        
        # Warning message if the number of nodes is not what it should be. Possible reasons: the energy added at the end
        #of the loop is too large; accidental zeros at the large r  end of the r segment are being counted (plot the det).
        
        if nodes != n-1:
            print('Incorrect number of nodes')
        
        # nu is the parameter of the linear convination of solutions that gives wave functions which converge at large r
        nu = -psi_p.y[0][-1]/psi_m.y[0][-1]
        
        # Creating the two wave functions and normalizing them
        norm = np.sqrt(simpson((psi_p.y[0]+nu*psi_m.y[0])**2+(psi_p.y[2]+nu*psi_m.y[2])**2,psi_p.t)) # we normalize the r.w.f
        npsi_sigma =(psi_p.y[0]+nu*psi_m.y[0])/norm
        npsi_pi =(psi_p.y[2]+nu*psi_m.y[2])/norm
         
        # Computing <1/r> and <E_kinetic>
        oor = simpson((npsi_sigma**2+npsi_pi**2)/psi_p.t,psi_p.t)
        kin = en - simpson((npsi_sigma**2*pot_Sigma(psi_p.t)+npsi_pi**2*pot_Pi(psi_p.t)),psi_p.t) 
        
        # Saving the wave fucntion in files
        np.savetxt(sec+'/wf_'+sec+'_'+coupling_label+'_'+str(l)+str(n)+'.txt', np.column_stack((psi_p.t, npsi_sigma, npsi_pi)), fmt='% 6.5f', delimiter='  ') # save r.w.f
        
        # Saving the states properties to dump in a file later
        spec.append([coupling_label,l, n, en, en + 2*mq, oor, kin]) # store spectrum data
    
        # Printing state properties and wave fucntions for checks
        print('l: {:1d},'.format(l),
              'n: {:1d},'.format(n),
              'nodes: {:1d},'.format(nodes),
              'E_b: {:.3f},'.format(en),
              'mass: {:.3f},'.format(2*mq+en),
              '<1/r>: {:.3f},'.format(oor),
              'E_kin: {:.3f}'.format(kin))
        
        plt.plot(psi_p.t*hbc, npsi_sigma, psi_p.t*hbc, npsi_pi)
        plt.axhline(y=0, color='black', linestyle='--')     
        plt.show()
        
        
# Uncomment to plot the determimant

#        plt.plot(psi_p.t[:-1]*hbc, det_psi[:-1])               
#        plt.ylim(-10,10)
#        plt.axhline(y=0, color='black', linestyle='--')     
#        plt.show()
        
        # Preparing to obtain the next state in the spectrum
        en = en + np.absolute(en)*0.01 # Jump the energy past the eigenvalue so the next loop looks for a higher energy one
        sf = 1 # reset the scale factor for the next loop

# Uncomment the following lines to save the spectrum in a file if the next cell is not run

# aux = pd.DataFrame(spec, columns =  ['coup','l','n','E0[GeV]','M[GeV] ','<1/r>  ','Ek[GeV]'])
# aux.to_csv(sec+'/spec'+sec+'.txt', sep = '\t', float_format='% .4f', index=False)  # save spectrum data into a file
```

Now we solve for the uncoupled cases. This essentially the same code used for conventional quarkonium. We have two cases: 
1) The above system of coupled equations for l=0 decouples as the off-diagonal and the $\Pi_u$ terms vanish.Howver note the centrifual term still exists due to the gluelump spin contribution.
2) The uncoupled eq. for the $\Pi_u$ state. See eq.52 in arXiv: 1510.04299. Note that, again, l>=1.
We use the incedental fact that the centrifual term in 1) is equal to the one for 2) for l=1.

```python
mq = sctsl(mc, mb) # picks heavy quark mass 

# Each step in the loop increses the energy tested by stp*10**(-sf)

stp = 1 # order 1 factor multipling the energy step
sf = 1  # scale factor controling the order of magnitude of the initial steps (now set to 0.1)
sd = 8  # significant digits in the final results of the energies (more or less...). More is needed for larger maxr
# spec = [] # Uncomment to run to this cell alone and not after the previous one.

for l_aux in range(0,3): # We loop over angular momentum 0,1,2; to be adjusted as needed
    
    if l_aux == 0:
        pot = pot_Sigma
        l=1 # Note that the eq. for uncoupled Pi state with l=1 is equivalent to l=0 coupled Sigma state
        coupling_label = 'p' 
    else:
        pot = pot_Pi
        l = l_aux
        coupling_label = 'm' 
    
    # p and m labelel the solutions corresponding to C21 and C22 of arXiv: 1510.04299, respectively. 
    # p=plus, m=minus which is the relative sign of the two \Pi terms with the r_+ definition we are using here
    
    en = 0.9    # Starting energy for the scan
    
    for n in range(1,6):  # We loop over principal quarntum number n=1..8; to be adjusted as needed
        
        # Minimum value of the range of r for which we will solve the Schroedinger eq. Must be larger than r of the first data 
        # point of the potential.
            
        minr = 10**(-4+l/4)
        
        # Maximum value of the range of r. Ideally the larger the better, however larg values result in the need to find the
        # eigenenergies with unpractical accuracies. The most practial max r is just a few tens of a fm beyond where the wave 
        # function becomes almost zero, which I determine by trial an error and cheking the resulting wave functions. For 
        # increasing n and l we expect the extension of the wave function to increase. However, the scalings are also obtained  
        # by trial an error (maybe exact solutions of linear potentials could be used as a guide too?).
        
        maxr = sctsl(2.05*(1.12**(n-1))*(1.15**l)/hbc,1.5*(1.1**(n-1))*(1.08**l)/hbc)               
        
        r_range =  [minr, maxr]
        
        # Boundary conditions of the r.w.f. and its derivative. These are obtained from the solution for the schroedinger
        # eq. in the short distance
        
        y_minr = [minr**(l+1),(l+1)*minr**l]
        
        # We rewrite the schroedinger eq. into a linear system y'=func(r,y)
        
        def func(r,y):
            return [y[1], (mq*(pot(r)-en)+l*(l+1)/r**2)*y[0]]   # y[0]= psi(r) (*reduced* w.f.) y[1]= psi(r)'
                 
        # we solve the diff. eq. for the initial value and find the sign of the divergence
        
        in_psi = solve_ivp(func,r_range,y_minr, rtol=10**(-sd-1), atol=10**(-sd-1)) 
        ref_sgn = np.sign(in_psi.y[0][-1]) 
        
        # Now we start the loop that scans the energy (from the initial value and incresing) until it finds an eigenenergy up to
        # some accuracy
        
        while True:
            en = en + stp*10**(-sf) # Adding energy step
            psi = solve_ivp(func,r_range,y_minr, rtol=10**(-sd-1), atol=10**(-sd-1)) # Solve diff.eq. for the energy of the step
            ref = np.absolute(psi.y[0][-2])/np.nanmax(np.absolute(psi.y[0])) # value of the r.w.f. at the edge of the range over it max.value
            sgn = np.sign(psi.y[0][-2]) # sign of the r.w.f at the edge of the r range

            if ref < 1e-3: # We approximate the eigenenergy by the value of en when ref is this small
                break      # We exit the loop
            if sf == sd:   # We exit the loop if the number of significat digits of the energy becomes too large
                print('too much!') # if this prints the convergences conditions are too harsh
                break
            if sgn != ref_sgn: # If the sign of the large r divergence changes means we have crossed an eigenvalue
                en = en - stp*10**(-sf) # Undo the last energy step
                sf=sf+1 # We increase the scale factor so the next energy step is one order of magnitude smaller.
        
        # we normalize the r.w.f
        norm = np.sqrt(simpson(psi.y[0]**2,psi.t))
        npsi = psi.y[0]/norm
        
        # We compute the number of nodes by counting adjacent points in the determinant with different sign
        nodes = np.sum(npsi[0:-11]*npsi[1:-10] < 0)
        # warning message if the number of nodes is not what it should be
        if nodes != n-1:
            print('Incorrect number of nodes')
         
        # Computing <1/r> and <E_kinetic>
        oor = simpson(npsi**2/psi.t,psi.t) 
        kin = en - simpson(npsi**2*pot(psi.t),psi.t) 
        
        # Saving the wave fucntion in files
        np.savetxt(sec+'/wf_'+sec+'_'+coupling_label+'_'+str(l_aux)+str(n)+'.txt', np.column_stack((psi.t, npsi)), fmt='% 6.5f', delimiter='  ') # save r.w.f
        
        # Saving the states properties to dump in a file later
        spec.append([coupling_label,l_aux, n, en, en + 2*mq, oor, kin]) 
        
        # Printing state properties and wave fucntions for checks
        print('l: {:1d},'.format(l_aux),
              'n: {:1d},'.format(n),
              'nodes: {:1d},'.format(nodes),
              'E_b: {:.3f},'.format(en),
              'mass: {:.3f},'.format(2*mq+en),
              '<1/r>: {:.3f},'.format(oor),
              'E_kin: {:.3f}'.format(kin))
        
        plt.plot(psi.t*hbc,npsi)
        plt.axhline(y=0, color='black', linestyle='--')  
        plt.show()
        
        # Preparing to obtain the next state in the spectrum
        en = en + np.absolute(en)*0.01 # Jump the energy past the eigenvalue so the next loop looks for a higher energy one
        sf = 1 # reset the scale factor for the next loop

# Saving the state data in a file        
aux = pd.DataFrame(spec, columns =  ['coup','l','n','E0[GeV]','M[GeV] ','<1/r>  ','Ek[GeV]'])
aux.to_csv(sec+'/spec'+sec+'.txt', sep = '\t', float_format='% .4f', index=False)  # save spectrum data into a file
```
